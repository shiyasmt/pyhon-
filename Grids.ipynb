```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.PairGrid(iris)
plt.show()
```

```python
A=sns.PairGrid(iris)
A.map(plt.scatter)
plt.show()
```

```python
A.map_diag(plt.hist)
A.map_upper(plt.scatter)
A.map_lower(sns.kdeplot)
plt.show()
```

```python
sns.pairplot(iris)
plt.show()
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
plt.show()
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
B=sns.FacetGrid(tips,col='time',row='smoker')
plt.grid(True)
B
plt.show()
```

```python
B.map(plt.hist,'total_bill')
plt.show()
```

```python
B=sns.FacetGrid(tips,col='time',row='smoker',hue='sex')
B=B.map(plt.scatter,'total_bill','tip').add_legend()
plt.show()
```

```python
C=sns.JointGrid(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
C.plot(sns.regplot,sns.distplot)
plt.show()
```
