```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'])
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':100})
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='sex',col='time')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day')
plt.show()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='day',col='time',palette='coolwarm',scatter_kws={'s':100})
plt.show()
```

```python
A=sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',aspect=0.6,height=8,hue='sex',markers=['o','v'],scatter_kws={'s':150})
for ax in A.axes.flat:
    ax.grid(True)
plt.show()
```
