```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
flights=sns.load_dataset('flights')
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tips.head()
```

```python
tips.corr(numeric_only=True)
```

```python
sns.heatmap(tips.corr(numeric_only=True))
plt.show()
```

```python
sns.heatmap(tips.corr(numeric_only=True),cmap='coolwarm',annot=True)
plt.show()
```

```python
pvflights=flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
pvflights
```

```python
sns.heatmap(pvflights)
plt.show()
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='white',linewidths=1)
plt.show()
```

```python
sns.heatmap(pvflights,cmap='magma',linecolor='w',linewidths=5)
plt.show()
```

```python
sns.clustermap(pvflights)
plt.show()
```

```python
sns.clustermap(pvflights,cmap='rainbow',standard_scale=1,linewidths=1,linecolor='black',annot=True)
plt.show()
```
