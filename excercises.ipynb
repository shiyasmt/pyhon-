```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
%matplotlib inline
```

# exercise 1

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Graph")
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlim([0,100])
axes.set_ylim([0,200])
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("Graph")
plt.yticks(np.arange(0,max(y)+50,50))
plt.show()
```

# exercise 2

```python
fig = plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.6,0.2,0.2])

axes1.plot(x,y)
axes1.set_xlabel('x axis for Graph A')
axes1.set_ylabel('y axis for Graph A')
axes1.set_title('Graph A')
axes2.plot(y,x)
axes2.set_xlabel('x axis for Graph B')
axes2.set_ylabel('y axis for Graph B')
plt.show() 
```

```python
fig = plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.6,0.2,0.2])

axes1.plot(x,y,'r')

axes1.set_xlabel('x axis for Graph A')
axes1.set_ylabel('y axis for Graph A')
axes1.set_title('Graph A')
axes2.set_xlim([0,100])
axes2.set_ylim([0,200])
axes1.set_xlim([0,100])
axes1.set_ylim([0,200])
axes2.plot(x,y,'r')
axes2.set_xlabel('x axis for Graph B')
axes2.set_ylabel('y axis for Graph B')
axes2.set_xticks(np.arange(0,max(x)+20,20))
axes2.set_yticks(np.arange(0,max(y)+50,50))
plt.show() 
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.plot(x, z)
ax1.set_xlabel("x axis for A")
ax1.set_ylabel("y axis for A")
ax1.set_title("graph A")
ax2.plot(y, x)
ax2.set_xlabel("x axis for B")
ax2.set_ylabel("y axis for B")
ax2.set_title("graph B")
ax2.set_xlim(20, 22)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.plot(x,z)
ax1.set_xlabel("x axis for A")
ax1.set_ylabel("y axis for A")
ax1.set_title("graph A")
ax1.set_ylim(0,10000)
ax1.set_xlim(0,100)
ax2.plot(x,y)
ax2.set_xlabel("x axis for B")
ax2.set_ylabel("y axis for B")
ax2.set_title("graph B")
ax2.set_xlim(20, 22)
ax2.set_ylim(30,50)
plt.show()
```

```python
fig,ax= plt.subplots(nrows=1, ncols=2) 
plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(nrows = 1, ncols = 2)
plt.tight_layout()

ax[0].plot(x, y,linewidth=5,
           linestyle='-',alpha=0.1,marker='o',
           markersize=20,markerfacecolor='yellow',
           markeredgewidth=5,markeredgecolor='green')
ax[1].plot(x, z, 'v')
plt.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2)
ax[0].plot(x,y,color='b',linestyle='--',lw=2)
ax[1].plot(x,z,'r')
ax[1].set_xlim(0,100)
ax[1].set_ylim(0,10000)
ax[0].set_ylim(0,200)
ax[0].set_xlim(0,100)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2,figsize=(9,3))
plt.tight_layout()
ax[0].plot(x, y,linewidth=5,
           linestyle='-',alpha=0.1,marker='o',
           markersize=15,markerfacecolor='yellow',
           markeredgewidth=5,markeredgecolor='green')
ax[1].plot(x, z, 'v')
plt.show()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,2))
ax[0].plot(x,y,color='b',linestyle='-',lw=4)
ax[1].plot(x,z,color='r',linestyle='--',lw=2)
ax[1].set_xlim(0,100)
ax[1].set_ylim(0,10000)
ax[0].set_ylim(0,200)
ax[0].set_xlim(0,100)
plt.tight_layout()
plt.show()
```
