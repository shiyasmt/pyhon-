```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.show()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
plt.show()
```

```python
tips['sex'].value_counts()
```

```python
sns.countplot(x='sex',data=tips)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,palette="rainbow")
plt.show()
```

```python
sns.boxplot(data=tips,palette='rainbow',orient='h')
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker",data=tips, palette="coolwarm")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,palette='Set1',hue="sex")
plt.show()
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,split=True,palette='Set1',hue="sex")
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False)
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=False,hue='sex')
plt.show()
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,jitter=True,hue='sex',palette='Set1',dodge=True)
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", dodge=True)
plt.show()
```

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='violin')
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='box')
plt.show()
```
